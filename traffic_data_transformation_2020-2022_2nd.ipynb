```python
import pandas as pd
import os
from itables import show
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

# Import csv's

```python
# Directory containing the CSV files
csv_directory = r"D:\Data Science\VS Code\final_project\Verkehrsdataen_Darmstadt\Daten_Verkehrsrechner_2020-2022"

# List all CSV files in the directory
csv_files = [f for f in os.listdir(csv_directory) if f.endswith('.csv')]
```

```python
# Initialize an empty DataFrame to store the merged data
merged_data = pd.DataFrame()
dfs = []
# Loop through the CSV files and concatenate them
for csv_file in csv_files:
    file_path = os.path.join(csv_directory, csv_file)
    df = pd.read_csv(file_path, sep=';')
    df['intersection'] = csv_file[0:4]
    dfs.append(df)

# Concatenate the DataFrames vertically (stack them on top of each other)
merged_data = pd.concat(dfs, axis=0, ignore_index=True, join = 'outer')

# Reset the index of the merged DataFrame
merged_data.reset_index(drop=True, inplace=True)
show(merged_data)
```

# Clean Traffic Dataframe

```python
#drop all columns about detector occupancy ('Vwz')
pattern = r'^Vwz_'


bel_df = merged_data.drop(columns=[col for col in merged_data.columns if pd.Series(col).str.contains(pattern).any()])
show(bel_df)
```

```python
#rename timestamp column
bel_df.columns = ['timestamp'] + list(bel_df.columns[1:])
bel_df
```

```python
#filter for detector columns

intersections = bel_df.pop('intersection')
timestamp = bel_df.pop('timestamp')

pattern = r'Bel_D'


bel_d_df = bel_df.filter(regex=pattern)
bel_d_df
```

```python
#readd intersections name and timestamp
bel_d_df['intersections'] = intersections
bel_d_df['timestamp'] = timestamp
bel_d_df
```

```python
#rename columns so they match the detectors df
bel_d_df.columns = bel_d_df.columns.str.replace('^Bel_', '', regex=True)
bel_d_df
```

```python
#drop all columns that contain only NaNs
bel_d_df.replace(np.nan, None, inplace=True)
bel_d_df.dropna(axis=1, how='all', inplace=True)

bel_d_df.fillna(0, inplace=True)

bel_d_df
```

```python
#convert timestamps index into datetime
bel_d_df.index = pd.to_datetime(bel_d_df['timestamp'], format='%d.%m.%Y %H:%M')
bel_d_df
```

```python
#drop old timestamp column
bel_d_df.drop(columns = 'timestamp', inplace=True)
bel_d_df
```

```python
#group by intersection and timestamp
bel_d_df_group = bel_d_df.groupby(['intersections', 'timestamp']).sum()
bel_d_df_group
```

```python
#convert object columns to numeric
bel_d_df.iloc[:, 0:42] = bel_d_df.iloc[:, 0:42].apply(pd.to_numeric, errors='coerce')
```

```python
#replaces all negative values in the df with 0

bel_d_df.iloc[:, 0:42] = bel_d_df.iloc[:, 0:42].applymap(lambda x: max(0, x))
```

```python
bel_d_df['hour'] = bel_d_df.index.hour
bel_d_df
```

```python
#replace outliers with mean of that hour for that respective intersections


for intersection_hour in bel_d_df[['intersections', 'hour']].drop_duplicates().itertuples(index=False):
    intersection = intersection_hour[0]
    hour = intersection_hour[1]
    
    #filter the DataFrame for the current intersection and hour
    condition = (bel_d_df['intersections'] == intersection) & (bel_d_df['hour'] == hour)
    filtered_df = bel_d_df[condition]
    
    #calculate the mean of each column for this subset
    mean_values = filtered_df.mean(numeric_only=True)
    
    #replace values above 3000 with the respective column's mean
    columns_to_replace = filtered_df.columns.difference(['intersections', 'hour'])
    bel_d_df.loc[condition, columns_to_replace] = bel_d_df.loc[condition, columns_to_replace].apply(lambda x: x.mask(x >= 3000, mean_values[x.name]))


show(bel_d_df)
```

```python
bel_d_df.drop(columns = 'hour', inplace=True)
bel_d_df
```

```python
#filter for inbound and outbound traffic
for intersection in detectors.intersection.unique():
    detectors_int = detectors.loc[detectors['intersection'] == intersection]
    print(intersection)
    for detector in detectors_int.detector:
        if detectors_int.loc[detectors_int['detector'] == detector, 'inbound'].any() == 1:
            print(detector)
            result.loc[(result['intersections'] == intersection), 
                               detector] *= -1
        elif detectors_int.loc[detectors_int['detector'] == detector, 'inbound'].any() == 0:
            print(detector)
            result
            
    
result
```

```python
# Apply percentages
for intersection in detectors.intersection.unique():
    detectors_int = detectors.loc[detectors['intersection'] == intersection]
    print(intersection)
    
    for detector in detectors_int.detector:
        if detectors_int.loc[detectors_int['detector'] == detector, 'inbound'].any() == 1:
            print(detector)
            detectors_int_per = detectors_int.loc[((detectors_int['detector'] == detector) & (detectors_int['inbound'] == 1))]
            percentage = detectors_int_per.loc[detectors_int_per['detector'] == detector, 'percentage'].item()
            
            # Update the values in the DataFrame using .loc
            result.loc[result['intersections'] == intersection, detector] *= percentage
        elif detectors_int.loc[detectors_int['detector'] == detector, 'inbound'].any() == 0:
            print(detector)
            detectors_int_per = detectors_int.loc[((detectors_int['detector'] == detector) & (detectors_int['inbound'] == 0))]
            percentage = detectors_int_per.loc[detectors_int_per['detector'] == detector, 'percentage'].item()
            
            # Update the values in the DataFrame using .loc
            result.loc[result['intersections'] == intersection, detector] *=  percentage
```

```python
# replace all not relevant values with 0 
for intersection in detectors.intersection.unique():
    detectors_int = detectors.loc[detectors['intersection'] == intersection]
    print(intersection)
    
    # Get a list of columns to exclude from being set to 0
    exclude_columns = detectors_int.detector.tolist()
    exclude_columns.append('intersections')
    
    # Get a list of columns to set to 0
    selected_columns = [col for col in result.columns if col not in exclude_columns]
    
    # Replace values with 0 for the specific intersection and selected columns
    result.loc[result['intersections'] == intersection, selected_columns] = 0


result
```

```python
#copy df
result_before_filtering = result.copy()
```

```python
cardinal_points_dict  = {'west' : ['A001', 'A126', 'A160'], 'east' : ['A027', 'A036'], 'north' : ['A111', 'A033', 'A069', 'A090', 'A010', 'A011', 'A008'], 'south': ['A020', 'A015', 'A067']}
```

```python
result_before_filtering['cardinal_point'] = ''
for cardinal, values in cardinal_points_dict.items():
    result_before_filtering['cardinal_point'] = result_before_filtering.apply(lambda row: cardinal if any(val in row['intersections'] for val in values) else row['cardinal_point'], axis=1)
```

```python
result_before_filtering.loc[result_before_filtering['intersections'] == 'A008']
```

```python
# Sum positive values along each row and store in a 'Positive_Sum' column
result_before_filtering['Positive_Sum'] = result_before_filtering.iloc[:, 0:42].apply(lambda row: row[row > 0].sum(), axis=1)

# Sum negative values along each row and store in a 'Negative_Sum' column
result_before_filtering['Negative_Sum'] = result_before_filtering.iloc[:, 0:42].apply(lambda row: row[row < 0].sum(), axis=1)

result_before_filtering.head()
```

```python
#group results by time and cardinal point
result_traffic2 = result_before_filtering.groupby([result_before_filtering.index, 'cardinal_point'])[['Negative_Sum', 'Positive_Sum']].sum()
result_traffic2['Negative_Sum'] = result_traffic2['Negative_Sum'] * (-1)
result_traffic2.rename(columns = {'Negative_Sum' : 'inbound', 'Positive_Sum': 'outbound'}, inplace=True)
result_traffic2 = result_traffic2.reset_index()
result_traffic2
```

```python
pivoted_df = result_traffic2.pivot(index='timestamp', columns='cardinal_point', values=['inbound', 'outbound'])

# Flatten the multi-level column index
pivoted_df.columns = [f"{col[1]} {col[0]}" for col in pivoted_df.columns]

# Reset the index to make 'date' a regular column
pivoted_df.reset_index(inplace=True)
pivoted_df
```

```python
pivoted_df.to_csv('traffic_2020-2022_orientation.csv', index= False)
```
